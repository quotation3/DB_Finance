### ngram 날짜별로 합치기

```
import pandas as pd
from tqdm import tqdm
```

```
news = pd.read_csv('/content/drive/MyDrive/DB_Finance/data/test_ngram_only_topic.csv')
news
```

```
for i in range(len(news)):
    try:
        news['nouns'][i] = ''.join(news['nouns'][i].replace('[','').replace(']','').replace("'",''))
        news['bigram'][i] = ''.join(news['bigram'][i].replace('[','').replace(']','').replace("'",''))
        news['trigram'][i] = ''.join(news['trigram'][i].replace('[','').replace(']','').replace("'",''))
        news['fourgram'][i] = ''.join(news['fourgram'][i].replace('[','').replace(']','').replace("'",''))
        news['fivegram'][i] = ''.join(news['fivegram'][i].replace('[','').replace(']','').replace("'",''))
    except:
        pass
```

```
news['ngram'] = ''
for i in range(len(news)):
    try:
        news['ngram'][i] = news['nouns'][i] +', '+ news['bigram'][i] +', '+ news['trigram'][i] +', '+ news['fourgram'][i] +', '+ news['fivegram'][i]
    except:
        pass
```

```
for i in range(len(news)):
    try:
        news['nouns'][i] = news['nouns'][i].split(',')
        news['ngram'][i] = news['ngram'][i].split(',')
    except:
        pass
```

```
news
```

```
news = news.sort_values(by="date")
news.reset_index(inplace=True, drop=True)
```

```
ngram_news = news[['date','ngram']]
```

```
ngram_news
```

```
ngram_news_by_date = ngram_news.groupby(ngram_news['date']).sum()
```

```
ngram_news_by_date.reset_index(inplace=True)
```

```
ngram_news_by_date
```

### 주식 개장 기준으로 날짜 변경

```
stock = pd.read_csv('/content/drive/MyDrive/DB_Finance/data/lg_chemical_stock.csv')
stock
```

```
news_by_date_stock = pd.merge(ngram_news_by_date,stock, how='outer', on='date')
```

```
news_by_date_stock["date"] = pd.to_datetime(news_by_date_stock["date"])
news_by_date_stock = news_by_date_stock.sort_values(by="date")
```

```
from math import isnan
for i in range(len(news_by_date_stock)):
    if isnan(news_by_date_stock['price'][i]):
        for j in range(i+1,len(news_by_date_stock)):
            if isnan(news_by_date_stock['price'][j]):
                continue
            else:
                news_by_date_stock['price'][i]=news_by_date_stock['price'][j]
                news_by_date_stock['volume'][i]=news_by_date_stock['volume'][j]
                news_by_date_stock['change'][i]=news_by_date_stock['change'][j]
                news_by_date_stock['date'][i]=news_by_date_stock['date'][j]
                break
```

```
news_by_date_stock = news_by_date_stock[['date','ngram']]
```

```
news_by_date_stock_only_market = news_by_date_stock.groupby(news_by_date_stock['date']).sum()
```

```
news_by_date_stock_only_market
```

### 날짜별 tone 계산

```
ps = pd.read_csv('/content/drive/MyDrive/DB_Finance/data/polarity_score_only_topic.csv')
ps.columns=['word','score']
ps
```

```
ps['label'] = ''
for i in range(len(ps)):
    try:
        if ps['score'][i] > 1:
            ps['label'][i] = 1
        elif ps['score'][i] < 1:
            ps['label'][i] = -1
    except:
        pass
```

```
ps.reset_index(inplace=True, drop=True)
```

```
news_by_date_stock_only_market['tone']=''
news_by_date_stock_only_market['word_cnt']=''
news_by_date_stock_only_market['polar_cnt']=''
```

```
for i in tqdm(range(len(news_by_date_stock_only_market))):
    cnt=0
    polar=0
    for j in range(len(ps)):
        try:
            if ps['word'][j] in news_by_date_stock_only_market['ngram'][i]:
                cnt+=1
                polar+=ps['label'][j]
            news_by_date_stock_only_market['word_cnt'][i] = cnt
            news_by_date_stock_only_market['polar_cnt'][i] = polar
            news_by_date_stock_only_market['tone'][i] = news_by_date_stock_only_market['polar_cnt'][i]/news_by_date_stock_only_market['word_cnt'][i]
        except:
            news_by_date_stock_only_market['tone'][i] = 0
```

```
news_by_date_stock_only_market
```

```
news_by_date_stock_only_market.reset_index(inplace=True)
```

```
news_by_date_stock_only_market["date"] = pd.to_datetime(news_by_date_stock_only_market["date"])
stock["date"] = pd.to_datetime(stock["date"])
```

```
final = pd.merge(news_by_date_stock_only_market,stock, how='inner', on='date')
```

```
final
```

### 상관관계 도출
