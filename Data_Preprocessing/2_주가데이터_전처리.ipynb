```
import pandas as pd
from datetime import datetime
```

```
stock = pd.read_csv('/content/drive/MyDrive/DB_Finance/data/lg_chemical_stock.csv', encoding='cp949')
stock
```

```
stock = stock[['일자','종가','거래량','등락률']]
stock.columns = ['date','price','volume','change']
```

```
stock['date'] = pd.to_datetime(stock['date'], format = "%Y/%m/%d").dt.strftime('%Y-%m-%d')
```

```
stock
```

```
stock.to_csv('/content/drive/MyDrive/DB_Finance/data/lg_chemical_stock.csv', index=False)
```
